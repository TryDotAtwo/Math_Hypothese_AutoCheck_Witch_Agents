Этот ноутбук показывает базовый рабочий процесс: генерация графов, вычисление инвариантов, проверка гипотезы.

```python
import sys
sys.path.insert(0, '..')

import networkx as nx
import matplotlib.pyplot as plt
from src.graphs import common_families, random_graph
from src.invariants import compute_invariants, INVARIANT_REGISTRY
from src.analysis import test_hypothesis
from src.visualization import draw_graph, plot_invariant_scatter
```

## 1. Генерация графов

```python
families = common_families(max_n=8)
print("Доступные семейства:")
for name, graphs in families.items():
    print(f"  {name}: {len(graphs)} графов")
```

## 2. Вычисление инвариантов

```python
G = nx.petersen_graph()
inv = compute_invariants(G)
print("Инварианты графа Петерсена:")
for key, val in inv.items():
    print(f"  {key} = {val}")
```

## 3. Проверка гипотезы

**Пример:** Для любого графа $G$, хроматическое число $\chi(G) \leq \Delta(G) + 1$ (теорема Брукса — с оговорками).

```python
def brooks_bound(G, inv):
    """chi(G) <= Delta(G) + 1"""
    return inv['chi'] <= inv['Delta'] + 1

result = test_hypothesis(brooks_bound, invariants=['chi', 'Delta', 'n', 'm'])
print(f"\nВсего графов: {result['total']}")
print(f"Контрпримеров: {len(result['counterexamples'])}")
```

## 4. Визуализация

```python
fig = draw_graph(nx.petersen_graph(), title="Граф Петерсена")
plt.show()
```

```python
# Scatter: хроматическое число vs максимальная степень
all_graphs = []
for family_graphs in families.values():
    all_graphs.extend(family_graphs)

data = [compute_invariants(G, ['chi', 'Delta']) for G in all_graphs]
fig = plot_invariant_scatter(data, 'Delta', 'chi', title='χ(G) vs Δ(G)')
plt.show()
```
